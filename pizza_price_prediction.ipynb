```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('pizza_data.csv')
```

```python
x = data[['Size (in inches)']]  # Independent variable (Pizza Size)
y = data['price (in $)']        # Dependent variable (Price)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
predicted_prices = model.predict(x)
```

```python
print("Model Coefficient (Slope):", model.coef_[0])
print("Model Intercept:", model.intercept_)
print("R-squared Score:", model.score(x, y))
```

```python
plt.scatter(x, y, color='blue', label='Actual Prices')
plt.plot(x, predicted_prices, color='red', label='Predicted Prices')
plt.xlabel('Pizza Size (in inches)')
plt.ylabel('Price (in $)')
plt.title('Pizza Price Prediction')
plt.legend()
plt.show()
```

```python
size =int(input("Enter the size of pizza: "))
new_size = np.array([[size]])
predicted_price_new = model.predict(new_size)
print(f"Predicted price for {size}-inch pizza: ${predicted_price_new[0]:.2f}")
```
